```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import norm
import seaborn as sns

def get_pdf_probability(dataset, startrange, endrange):
    # Check if the 'ssc_p' column exists in the DataFrame
    if "ssc_p" not in dataset.columns:
        raise KeyError("Column 'ssc_p' not found in dataset")

    # Create the distribution plot for 'ssc_p'
    sns.histplot(dataset["ssc_p"], kde=True, color='red')

    # Add vertical lines for the start and end range
    plt.axvline(startrange, color='blue')
    plt.axvline(endrange, color='blue')

    # Calculate sample statistics
    sample = dataset["ssc_p"]
    sample_mean = sample.mean()
    sample_std = sample.std()

    # Create a normal distribution with the sample mean and standard deviation
    dist = norm(sample_mean, sample_std)

    # Calculate the probability for each value in the specified range
    values = range(startrange, endrange)
    probabilities = [dist.pdf(value) for value in values]

    # Sum of probabilities within the specified range
    prob = sum(probabilities)
    
    # Display the result
    print(f"THE AREA BETWEEN RANGE({startrange},{endrange}): {prob}")
    
    return prob

# Example usage:
# Assuming 'dataset' is your DataFrame loaded with data
get_pdf_probability(dataset, 60, 70)
```
